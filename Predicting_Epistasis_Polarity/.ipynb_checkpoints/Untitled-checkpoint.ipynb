Import interactors from the Yeastmine Database of genes that are present in saccharomyces cerevisiae

```python
import pandas as pd
import numpy as np
import csv


# Read in the Interactions of the genes that are present in Saccharomyces cereviae
Int_Sacc_Genes = pd.read_csv('Interactions_Saccharomyces/Network_Interactions_Sacc_Genes.tsv',sep='\t')

Int_Sacc_Genes.head()
```

```python
Int_Drf1 = pd.read_csv('Interactions_Saccharomyces/Network_Interactions_Drf1.tsv',sep='\t')

Int_Drf1.rename(columns={'Gene > Standard Name':Int_Sacc_Genes.columns[3],
                          'Gene > Systematic Name':Int_Sacc_Genes.columns[4],
                            'Gene > Primary DBID': Int_Sacc_Genes.columns[5]}, 
                 inplace=True)

Int_Drf1.head()
Int_Drf1.insert(0,"Gene > Standard Name","DRF1")

for x in Int_Sacc_Genes.columns:
    
    if Int_Drf1.columns.contains(x):
        print(x)
    else:
        Int_Drf1.insert(0,x,"NA")
        
Int_Drf1.head()
            
```
