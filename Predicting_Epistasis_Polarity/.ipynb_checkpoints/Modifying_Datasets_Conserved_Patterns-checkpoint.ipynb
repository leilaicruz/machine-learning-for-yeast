```python
import pandas as pd
import numpy as np

# Import the data as a single ExcelFile object. Different datasets are stored in different tabs of the excelfile 
#(Repeated Combination, Unique Combinations and Core Proteins)
Pol_Data = pd.ExcelFile('Datasets_Conserved_Patterns_Polarity_Proteins\Repertoire_Edited.xlsx')

# Extract the seperate datasets into different DataFrames
Rep_Comb = pd.read_excel(Pol_Data, 'Repeated Combinations')
Unique_Comb = pd.read_excel(Pol_Data, 'Unique Combinations')
Core = pd.read_excel(Pol_Data, 'Core Proteins')

#Create a list of all analyzed proteins and the proteins noted as 'Core proteins'
All_Proteins = list(Rep_Comb.columns[2:])
Core_Proteins = list(Core.columns[2:])

#Create a DataFrame of all protein combinations that occur (Repeated+Unique)
C_Rep = Rep_Comb.loc[:,All_Proteins[0]:All_Proteins[len(All_Proteins)-1]]
C_Unique = Unique_Comb.loc[:,All_Proteins[0]:All_Proteins[len(All_Proteins)-1]]
Protein_Comb_All = C_Rep.append(C_Unique) 

#Print the matrix containing all protein combinations that occur in nature
print(Protein_Comb_All)
```

```python
#Code that can be used to get the interactions between all known yeast genes from YeastMine

from __future__ import print_function
from intermine.webservice import Service
service = Service("https://yeastmine.yeastgenome.org/yeastmine/service")
query = service.new_query("Gene")
query.add_view(
    "secondaryIdentifier", "symbol", "interactions.participant2.symbol",
    "interactions.participant2.secondaryIdentifier"
)
query.add_sort_order("Gene.symbol", "ASC")
query.add_constraint("Gene", "IN", "ALL_Verified_Uncharacterized_Dubious_ORFs", code = "A")

for row in query.rows():
    print(row["secondaryIdentifier"], row["symbol"], row["interactions.participant2.symbol"], \
        row["interactions.participant2.secondaryIdentifier"])
```
